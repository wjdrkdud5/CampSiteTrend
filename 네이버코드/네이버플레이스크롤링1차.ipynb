```python
from turtle import clear
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

def campsite(camp):
    res = requests.get('https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=' + camp)
    html = res.content.decode('utf-8','replace')
    return html

result_list = [['name', 'price', 'rating', 'review_name', 'review_cnt']]

def cmp_dt(name_list): 
  L = len(name_list)
  i = 0
  for name in name_list:
    i += 1
    try:
      soup = BeautifulSoup(campsite(name), 'html.parser')
      review_text = soup.select("div._2B33i > span._10uv0")
      review_text_list = []
      for i in range(0,len(review_text)):
          review_text = soup.select("div._2B33i > span._10uv0")[i].text
          review_text_list.append(review_text.replace("\"", ""))
      review_num = soup.select("div._2B33i > span._3IFxe")
      review_num_list = []
      for i in range(0,len(review_num)):
          review_num = soup.select("div._2B33i > span._3IFxe")[i].text[13:]
          review_num_list.append(review_num)
      rate_num = soup.select("div._20Ivz > span._1Y6hi._1A8_M > em")[0].text
      price_num = soup.select("div._2QlZz > span.FSeNw")
      price_num_list = []
      price_total_list = ''
      for i in range(0,len(price_num)):
          price_num = soup.select("div._2QlZz > span.FSeNw")[i].text
          price_total_list = price_total_list + price_num
          if len(price_num)-1 != i:
              price_total_list = price_total_list + ", "
      price_num_list = ''
      for i in range(0, len(review_text_list)):
          price_num_list = price_num_list  + review_text_list[i] + "(" + review_num_list[i] + ")"
          if len(review_text_list)-1 != i:
              price_num_list = price_num_list + ", "
      review_total_list = ''
      for i in range(0, len(review_text_list)):
        result_list.append([name, price_total_list, rate_num, review_text_list[i], review_num_list[i]])
        
      result_list.append([name, price_total_list, rate_num, review_text_list, review_num_list])
    except:
      print('{}/{}--'.format(i, L),"정보없음--" + name)
      pass

  cmp_df = pd.DataFrame(result_list)
  return cmp_df





# f = open( 'num_blogs.csv','w', encoding = 'utf-8-sig', newline='')
# wr = csv.writer(f)
# for result in result_list:
#     wr.writerow(result)
# f.close()
```
