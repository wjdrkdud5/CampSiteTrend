##   =>   좌측 상단 네번째 폴더 누르고 로컬에 있는 데이터 업로드

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SparkSQL").getOrCreate()

import pyspark
```

```python
from pyspark.sql.types import StructType, StructField
from pyspark.sql.types import StringType, FloatType, IntegerType
Schema = StructType([
    StructField("div", StringType(), True),
    StructField("name", StringType(), True),
    StructField("price", IntegerType(), True),
    StructField("virev", IntegerType(), True),
    StructField("blgrev", IntegerType(), True),
    StructField("addr1", StringType(), True),
    StructField("addr2", StringType(), True),
    StructField("attr", StringType(), True),
    StructField("point", FloatType(), True),
    StructField("revatt", StringType(), True),
    StructField("revpnt", IntegerType(), True),
])
```

```python
datas  = spark.read.csv(f"/content/N_cmp_info1.csv", schema=Schema)
datas.show()
```

```python
datas  = spark.read.csv(f"/content/N_cmp_info2.csv", schema=Schema)
datas.show()
```
